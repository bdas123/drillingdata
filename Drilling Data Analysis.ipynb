```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_excel("drilling data.xlsx")
```

```python
df = dataframe.drop('Fig.5', axis = 1)
```

```python
df1 = df.iloc[0:20, :]
```

```python
df1 = df1.dropna()
```

```python
new_header = df1.iloc[0]
df1.columns = new_header
df1 = df1[1:]
```

```python
df2 = df.iloc[20:35]
new_header = df2.iloc[0]
df2.columns = new_header
df2 = df2[1:]
df2 = df2.dropna()
```

```python
df3 = df.iloc[35:46]
new_header = df3.iloc[0]
df3.columns = new_header
df3 = df3[1:]
df3 = df3.dropna()
```

```python
df4 = df.iloc[46:88]
new_header = df.iloc[0]
df4.columns = new_header
df4 = df4[1:]
df4 = df4.drop(['Lh-rpo'], axis = 1)
```

```python
import matplotlib.pyplot as plt
drilling_fluid_density = df1.iloc[:,0]
lh_fp = df1.iloc[:,1]
lh_rpr = df1.iloc[:,2]
lh_rpo = df1.iloc[:,3]
plt.plot(drilling_fluid_density, lh_fp,label = 'lh_fp') 
plt.plot(drilling_fluid_density, lh_rpr, label = 'lh_rpr')  
plt.plot(drilling_fluid_density, lh_rpo, label = 'lh_rpo')
plt.xlabel('Drilling Fluid Density (g/cm3)')
plt.ylabel('Lh')
plt.title('Lh vs Drilling Fluid Density')
plt.legend()
```

```python
drilling_fluid_flow_rate = df2.iloc[:, 0]
lh_fp = df2.iloc[:, 1]
lh_rpr = df2.iloc[:,2]
lh_rpo = df2.iloc[:,3]
plt.plot(drilling_fluid_flow_rate, lh_fp,label = 'lh_fp') 
plt.plot(drilling_fluid_flow_rate, lh_rpr, label = 'lh_rpr')  
plt.plot(drilling_fluid_flow_rate, lh_rpo, label = 'lh_rpo')
plt.xlabel('Drilling Fluid Flow Rate (L/sec)')
plt.ylabel('Lh')
plt.title('Lh vs Drilling Fluid Flow Rate')
plt.legend()
```
